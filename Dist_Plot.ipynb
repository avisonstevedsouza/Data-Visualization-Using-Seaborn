```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'/content/hr_data.csv')
df.head()
```

```python
sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

```python
df.describe()
```

# bins refer to the scale in x axis

```python
bins=[2,4,6,8,10]
sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

Removes KDE (Kernel Density Estimation)

```python
sns.distplot(df['time_spend_company'],bins=bins,kde=False)
```

Removes histogram

```python
sns.distplot(df['time_spend_company'],bins=bins,hist=False)
```

Increses the x axis dimensions

```python
sns.distplot(df['time_spend_company'],bins=bins,rug=True)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,rug=True)
plt.xticks(bins)
```

Histogram modification

```python
sns.distplot(df['time_spend_company'],
             bins=bins,
             rug=True,
             hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5})
plt.xticks(bins)
```

A KDE (Kernel Density Estimation) Modification

```python
sns.distplot(df['time_spend_company'],
             bins=bins,
             rug=True,
             hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5},
             kde_kws={'color':'black','linewidth':3,'alpha':0.5})
plt.xticks(bins)
```

Coloring whole graph

```python
sns.distplot(df['time_spend_company'],
             bins=bins,
             rug=True,
             color='green')
plt.xticks(bins)
```

Rotate graph

```python
sns.distplot(df['time_spend_company'],
             bins=bins,
             rug=True,
             color='green',
             vertical=True)
plt.xticks(bins)
```
