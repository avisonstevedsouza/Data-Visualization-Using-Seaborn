```
df=pd.read_csv(r'/content/hr_data.csv')
df.head()
```

```
sns.lineplot(x='number_project',y='average_montly_hours',data=df)
plt.title('Average Monthly hours for each project')
```

```
sns.lineplot(x='promotion_last_5years',y='left',data=df)
plt.title('Promotion vs Left')
```

```
sns.lineplot(x='department',y='left',data=df)
plt.figure(figsize=(120,160))
```

```
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,
             hue='left')
```

```
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,
             hue='department')
```

```
plt.figure(figsize=(12,6))
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,
             hue='left',
             style='department')
```

```
plt.figure(figsize=(12,6))
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,hue='left',
             style='department',
             legend=False)
```

```
plt.figure(figsize=(12,6))
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,hue='left',
             style='department',
             legend='full')
```

```
plt.figure(figsize=(12,6))
sns.lineplot(x='number_project',
             y='average_montly_hours',
             data=df,
             hue='left',
             style='department',
             legend='full',
             palette='flare')
```
